```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import ttest_ind, chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/insurance_data_cleaned.csv")
```

```python
df["HasClaim"] = np.where(df["TotalClaims"] > 0, 1, 0)
```

```python
df["Margin"] = df["TotalPremium"] - df["TotalClaims"]
```

```python
province_table = pd.crosstab(df["Province"], df["HasClaim"])

chi2, p, dof, expected = chi2_contingency(province_table)

print("P-value:", p)
```

```python
if p < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
zip1_margin = df[df["PostalCode"] == df["PostalCode"].unique()[0]]["Margin"]

zip2_margin = df[df["PostalCode"] == df["PostalCode"].unique()[1]]["Margin"]

t_stat, p = ttest_ind(zip1_margin, zip2_margin, equal_var=False)

print("P-value:", p)
```

```python
male_claims = df[df["Gender"] == "Male"]["TotalClaims"]

female_claims = df[df["Gender"] == "Female"]["TotalClaims"]

t_stat, p = ttest_ind(male_claims, female_claims, equal_var=False)

print("P-value:", p)
```

```python
# Hypothesis 1: Province Risk Difference
province_table = pd.crosstab(df["Province"], df["HasClaim"])

chi2, p1, dof, expected = chi2_contingency(province_table)

print("Province p-value:", p1)
```

```python
# Hypothesis 2: Zip Code Risk Difference

zip_claims = df[df["TotalClaims"] > 0]

zip1 = zip_claims[
    zip_claims["PostalCode"] == zip_claims["PostalCode"].unique()[0]
]["TotalClaims"]

zip2 = zip_claims[
    zip_claims["PostalCode"] == zip_claims["PostalCode"].unique()[1]
]["TotalClaims"]

t_stat, p2 = ttest_ind(zip1, zip2, equal_var=False)

print("Zip Risk p-value:", p2)
```

```python
# Hypothesis 3: Margin Difference

zip1_margin = df[
    df["PostalCode"] == df["PostalCode"].unique()[0]
]["Margin"]

zip2_margin = df[
    df["PostalCode"] == df["PostalCode"].unique()[1]
]["Margin"]

t_stat, p3 = ttest_ind(zip1_margin, zip2_margin, equal_var=False)

print("Margin p-value:", p3)
```

```python
# Hypothesis 4: Gender Risk Difference

male_claims = df[df["Gender"] == "Male"]["TotalClaims"]

female_claims = df[df["Gender"] == "Female"]["TotalClaims"]

t_stat, p4 = ttest_ind(
    male_claims,
    female_claims,
    equal_var=False
)

print("Gender p-value:", p4)
```

```python
results = pd.DataFrame({
    "Hypothesis": [
        "Province Risk Difference",
        "Zip Code Risk Difference",
        "Zip Code Margin Difference",
        "Gender Risk Difference"
    ],
    "Test": [
        "Chi-Square",
        "T-Test",
        "T-Test",
        "T-Test"
    ],
    "P-Value": [
        p1,
        p2,
        p3,
        p4
    ]
})

results["Decision"] = results["P-Value"].apply(
    lambda x: "Reject H0" if x < 0.05 else "Fail to Reject H0"
)

results
```

Province Risk

We reject the null hypothesis for provinces because the p-value is below 0.05. This indicates statistically significant differences in claim frequency across provinces. ACIS should consider region-based premium adjustments and targeted marketing strategies for lower-risk provinces.

Gender Risk

We fail to reject the null hypothesis for gender. This suggests that claim behavior between men and women is statistically similar, meaning gender-based pricing adjustments may not be justified.
